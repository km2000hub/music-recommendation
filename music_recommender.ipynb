```python
# to import pandas library 
import pandas as pd
#to implement DecisionTree algorithm 
from sklearn.tree import DecisionTreeClassifier
#split our data set into two data set
from sklearn.model_selection import train_test_split
#so you can use this function to calculate how accurate your model's predictions are by comparing them to the actual values.
from sklearn.metrics import accuracy_score
#to store our model - import joblib directly instead of from sklearn.externals
import joblib  # This is the corrected import statement
#to read the csv file , music_data= to name our data frame music_data
music_data=pd.read_csv('music.csv')
#to call our data frame
music_data
```

```python
#remove genre to set it as output and use X as input
X=music_data.drop(columns=['genre'])
#check if everything run as it should be
X
```

```python
#use Y to set it as output
Y=music_data['genre']
Y
#we only have the output here ( answer )
```

```python
#X_train , X_test input set for traingin and test Y_train , Y_test out put set for training and test
X_train , X_test ,Y_train , Y_test= train_test_split(X, Y, test_size=0.2)
```

```python
#model to train it 
model = DecisionTreeClassifier()
#takes the input set X and the output set Y
model.fit(X_train ,Y_train)
predictions=model.predict(X_test)
# This code calculates the accuracy of your machine learning model's predictions and then displays the result.
score = accuracy_score(Y_test, predictions)
score
```

```python
# to save your trained model to a file.
joblib.dump(model,'music-recommender.joblib')
```
